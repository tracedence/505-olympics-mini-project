```python
import pandas as pd
import numpy as np
 
```

```python
def load_data(file_path):
    
    data = pd.read_csv(file_path, skiprows = 1)
    
    df_csv = pd.DataFrame(data)
    
    #renaming columns
    df_csv.rename(columns= {'01 !':'Gold', '02 !':'Silver', '03 !':'Bronze','01 !.2':'Gold', '02 !.2':'Silver', '03 !.2':'Bronze','01 !.1':'Gold', '02 !.1':'Silver', '03 !.1':'Bronze'}, inplace = True)
    
    #split the unnamed column into two columns 
    df_csv[['Country name','Country code']] = pd.DataFrame(df_csv['Unnamed: 0'].str.split('\xa0').tolist() )
    
    df_csv.index = df_csv['Country name'] # change index to country name
    
    df_csv.drop(['Country name','Unnamed: 0', 'Total', 'Total.1'], axis = 1, inplace = True) # deleting country and Unnamed :0 columns
     
    
    #remove extra extra character
    df_csv['Country code'] = df_csv['Country code'].astype(str).apply(lambda x: x.split(" ",1)[0])
    
    df_csv.drop(['Totals'], axis = 0, inplace = True)
    
    
    return df_csv
  
```

```python
def first_country(df):
    
    return df.iloc[0]
    
```

```python
def gold_medal(df):
    
    """ take data frame then select gold column of summer games and after that sort values in descending order """
    
    return df['Gold'].iloc[:,0].sort_values(ascending = False).index[0]
```

```python
def biggest_difference(df):
    summer_gold = df['Gold'].iloc[:,0]
    winter_gold = df['Gold'].iloc[:,1].sort_values(ascending = False)
    return (abs(summer_gold -winter_gold)).sort_values(ascending = False).index[0]
```

```python
def get_points(df):
    gold = df['Gold'].iloc[:,0] + df['Gold'].iloc[:,1] + df['Gold'].iloc[:,2]
    silver = df['Silver'].iloc[:,0] + df['Silver'].iloc[:,1] + df['Silver'].iloc[:,2]
    bronze = df['Bronze'].iloc[:,0] + df['Bronze'].iloc[:,1] + df['Bronze'].iloc[:,2]
    
    total_points = (gold * 3 + silver *2 + bronze*1)
    return total_points.sort_values(ascending = False)
```

```python
def __main__():
    file_path = '../data/olympics.csv'
    display(load_data(file_path)) #for displaying whole dataframe
    df = load_data(file_path)
    print('Details of first country')
    print(first_country(df))
    
    print('country has won  most gold in summer games')
    print(gold_medal(df))
    
    print('biggest difference between their summner and winter gold medal count')
    print(biggest_difference(df))
    
    print('points')
    points  = get_points(df)
    df['points'] = points
    
    display(df)
```

```python
__main__()
```
