```python
import pandas as pd
import numpy as np
```

```python
def load_data(file_path):
    
    data = pd.read_csv(file_path, skiprows = 1)
    
    df_csv = pd.DataFrame(data)
    
    df_csv.rename(columns= {'01 !':'Gold', '02 !':'Silve', '03 !':'Bronze'}, inplace = True)
    
    #split the unnamed column into two columns 
    df_csv[['Country name','Country code']] = pd.DataFrame(df_csv['Unnamed: 0'].str.split('\xa0').tolist() )
    
    df_csv.index = df_csv['Country name'] # change index to country name
    
    df_csv.drop(['Country name','Unnamed: 0', 'Total'], axis = 1, inplace = True) # deleting country and Unnamed :0 columns
    
    
    # rearrange the columns 
    df_csv = df_csv[['Country code','# Summer', 'Gold', 'Silve', 'Bronze', '# Winter', '01 !.1',
       '02 !.1', '03 !.1', 'Total.1', '# Games', '01 !.2', '02 !.2', '03 !.2',
       'Combined total', ]]
    
    #remove extra extra character
    df_csv['Country code'] = df_csv['Country code'].astype(str).apply(lambda x: x.split(" ",1)[0])
    
    print(df_csv.columns)
    display(df_csv) #to print whole dataframe
  
```

```python
file_path = '../data/olympics.csv'
load_data(file_path)
```
